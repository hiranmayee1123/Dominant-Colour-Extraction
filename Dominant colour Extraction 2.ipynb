# Dominant Colour Extraction

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
img=cv2.imread('nature2.jpg')
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#cv2.imshow('image',img)
img.shape
```

```python
plt.imshow(new_img)
```

```python
#flatten all channels
pixels=new_img.reshape((-1,3))
print(pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5)
km.fit(pixels)
```

```python
#RGB values
centres=km.cluster_centers_
centres
centres=np.array(centres,dtype='uint8')
print(centres)
```

# Plot these channels

```python
i=1
plt.figure(0,figsize=(10,2))
colors=[]

for col in centres:
    plt.subplot(1,10,i)
    plt.axis('off')
    i=i+1
    colors.append(col)
    #create a color palatte
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=col
    plt.imshow(a)
    plt.show()
```
